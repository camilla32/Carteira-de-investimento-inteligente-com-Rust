```python
%%writefile main.rs
use std::fs::File;
use std::io::Write;

fn main() {

    let json = r#"
{
    "patrimonio": 10000,
    "rentabilidade": 12.8,
    "risco": 64,
    "ativos":[
        {
            "nome":"PETR4",
            "valor":5000
        },
        {
            "nome":"VALE3",
            "valor":3000
        },
        {
            "nome":"TESOURO",
            "valor":2000
        }
    ]
}
"#;

    let mut file =
        File::create("portfolio.json")
        .unwrap();

    file.write_all(json.as_bytes())
        .unwrap();

    println!("portfolio.json criado!");
}
```

```python
!$HOME/.cargo/bin/rustc main.rs
```

```python
!./main
```

```python
from IPython.display import HTML
import json

with open("portfolio.json") as f:
    data = json.load(f)

labels = [a["nome"] for a in data["ativos"]]
values = [a["valor"] for a in data["ativos"]]

HTML(f"""
<!DOCTYPE html>
<html>

<head>

<script src="https://cdn.jsdelivr.net/npm/chart.js"></script>

<style>

body{{
background:#0f172a;
color:white;
font-family:Arial;
padding:30px;
}}

.hero{{
text-align:center;
}}

.kpis{{
display:flex;
gap:20px;
margin-top:20px;
}}

.card{{
background:#1e293b;
padding:20px;
border-radius:10px;
flex:1;
}}

canvas{{
background:white;
border-radius:10px;
padding:10px;
margin-top:30px;
}}

</style>

</head>

<body>

<div class="hero">
<h1>Smart Investment Portfolio</h1>
<p>Dashboard alimentado por Rust + JSON</p>
</div>

<div class="kpis">

<div class="card">
<h3>Patrimônio</h3>
<h2>R$ {data["patrimonio"]:,.0f}</h2>
</div>

<div class="card">
<h3>Rentabilidade</h3>
<h2>{data["rentabilidade"]}%</h2>
</div>

<div class="card">
<h3>Score de Risco</h3>
<h2>{data["risco"]}</h2>
</div>

</div>

<canvas id="chart"></canvas>

<script>

new Chart(
document.getElementById('chart'),
{{
type:'pie',
data:{{
labels:{labels},
datasets:[{{
data:{values}
}}]
}}
}}
);

</script>

</body>

</html>
""")
```
